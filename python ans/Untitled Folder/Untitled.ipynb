```python
import re
```

```python
def IsInteger():
    inp = str(raw_input())
    mo = re.match('.*[0-9].*',inp)
    if mo:
        return True
    else:
        return False
```

```python
def IsFloat():
    inp = str(raw_input())
    mo = re.match('.*[0-9].[0-9].*',inp)
    if mo:
        return True
    else:
        return False
```

```python
def HasVowel():
    inp = str(raw_input())
    mo = re.match('.*[aAeEiIoOuU].*',inp)
    if mo:
        return True
    else:
        return False
```

```python
def IsHex():
    inp = str(raw_input())
    mo = re.match('[r'\s--([0-9a-fA-F]+)(?:--)?\s']',inp)
    if mo:
        return True
    else:
        return False
```

```python
def IsDate():
    inp = str(raw_input())
    mo = re.match('[0-9][0-9]-[0-9][0-9]-[0-9][0-9][0-9][0-9]',inp)
    if mo:
        return True
    else:
        return False
```

```python
def IsValidPassword():
    inp = str(raw_input())
    mo = re.match('(?=.*[a-z].*.*[a-z].*)(?=.*[A-Z].*.*[A-Z].*)(?=.*[0-9].*.*[0-9].*)\w{8,16}',inp)
#    mo = re.match('(?=.*[a-z].*[a-z].*)\w{8,16}',inp)
    if mo:
        return True
    else:
        return False
```

```python
def IsValidEmail():
    inp = str(raw_input())
    mo = re.match('[a-z0-9]*@[a-z]*.[c][o][m]',inp)
    if mo:
        return True
    else:
        return False
```
